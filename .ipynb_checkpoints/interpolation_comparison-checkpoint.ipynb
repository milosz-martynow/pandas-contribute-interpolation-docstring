```python
"""Acording to the documentation of interpolation.indetrp1d 
LINK: https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp1d.html
CITATION: ‘zero’, ‘slinear’, ‘quadratic’ and ‘cubic’ refer to a spline interpolation of zeroth, first, second or third order

we should expect that spline method with specified order = 1, will give the same results as slinear. 
In below simple example it is shown that such situation does not appear. 
It is clearly visible that  splinear with order=1 keeps "wavy" splinear behavur, 
whereas slinear exposes interpolation via linear function.

Two test data were prepared
Case 1. Gaps to interpolate in original y_all data are placed every second y array element resulting with y_gap_1.
Case 2. Gaps to interpolate in original y_all data are placed in random y array elements resulting with y_gap_1.

Those tests were performed on Scipy library, and on Pandas library, which uses Scipy.
"""
from scipy import interpolate
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import functools
from typing import Tuple

"""Propose data."""
x = [i for i in range(-10, 11, 1)]
y_all = [0, 0, 0, 0, 0, 0, 0, 0, 2, 5, 10, 5, 2, 0, 0, 0, 0, 0, 0, 0, 0,]
y_gap_1 = [0, np.nan, 0, np.nan, 0, np.nan, 0, np.nan, 2, np.nan, 10, np.nan, 2, np.nan, 0, np.nan, 0, np.nan, 0, np.nan, 0,]
y_gap_2 = [0, np.nan, 0, 0, np.nan, 0, 0, np.nan, np.nan, 5, 10, np.nan, 2, 0, np.nan, 0, 0, 0, np.nan, 0, 0,]

"""Preprocess data for both libraries."""
original = pd.Series(data=y_all, index=x, name="Original")
gaps_1 = pd.Series(data=y_gap_1, index=x, name="Original with gaps 1")
gaps_2 = pd.Series(data=y_gap_2, index=x, name="Original with gaps 2")

"""Functoions for both libraries."""
def plot(original: pd.Series,
         gaps_1_method_1: pd.Series, 
         gaps_1_method_2: pd.Series, 
         gaps_2_method_1: pd.Series, 
         gaps_2_method_2: pd.Series) -> Tuple[plt.Figure, plt.Axes, plt.Axes]:
    
    fig, (gaps_1, gaps_2) = plt.subplots(nrows=2, ncols=1, sharex=True)
    
    original.plot(ax=gaps_1)
    gaps_1_method_1.plot(ax=gaps_1)   
    gaps_1_method_2.plot(ax=gaps_1)
    gaps_1.legend()
    gaps_1.grid()
    gaps_1.set_ylabel("Gaps case 1 - y[-]")

    original.plot(ax=gaps_2)
    gaps_2_method_1.plot(ax=gaps_2)
    gaps_2_method_2.plot(ax=gaps_2)
    gaps_2.legend()
    gaps_2.grid()
    gaps_2.set_ylabel("Gaps case 2 - y[-]")
    gaps_2.set_xlabel("x[-]")

    fig.tight_layout()
    
    return fig, gaps_1, gaps_2
```

```python
"""Check Scipy interpolation behaviur."""
x_gap_1 = gaps_1.dropna().index.to_numpy()
y_gap_1 = gaps_1.dropna().to_numpy() 

x_gap_2 = gaps_2.dropna().index.to_numpy()
y_gap_2 = gaps_2.dropna().to_numpy() 

f_gaps_1_linear = interpolate.interp1d(x_gap_1, y_gap_1, kind="linear")
f_gaps_2_linear = interpolate.interp1d(x_gap_2, y_gap_2, kind="linear")
f_gaps_1_slinear = interpolate.interp1d(x_gap_1, y_gap_1, kind="slinear")
f_gaps_2_slinear = interpolate.interp1d(x_gap_2, y_gap_2, kind="slinear")

gaps_1_linear = f_gaps_1_linear(x)
gaps_2_linear = f_gaps_2_linear(x)
gaps_1_slinear = f_gaps_1_slinear(x)
gaps_2_slinear = f_gaps_2_slinear(x)

gaps_1_linear = pd.Series(data=gaps_1_linear, index=x)
gaps_2_linear = pd.Series(data=gaps_2_linear, index=x)
gaps_1_slinear = pd.Series(data=gaps_1_slinear, index=x)
gaps_2_slinear = pd.Series(data=gaps_2_slinear, index=x)

gaps_1_linear.name = "linear"
gaps_2_linear.name = "linear"
gaps_1_slinear.name = "slinear"
gaps_2_slinear.name = "slinear"

fig_scipy, scipy_gaps_1_ax, scipy_gaps_2_ax = plot(original=original,
    gaps_1_method_1=gaps_1_linear,
    gaps_1_method_2=gaps_1_slinear,
    gaps_2_method_1=gaps_2_linear,
    gaps_2_method_2=gaps_2_slinear)
fig_scipy.savefig("scipy_interpolation.png")
```

```python
"""Check Pandas behaviour, which uses Scipy interpolation module."""
gaps_1_spline = gaps_1.interpolate(method="spline", order=1)
gaps_1_spline.name = "spline, order 1"
gaps_2_spline = gaps_2.interpolate(method="spline", order=1)
gaps_2_spline.name = "spline, order 1"

gaps_1_slinear = gaps_1.interpolate(method="slinear")
gaps_1_slinear.name = "slinear"
gaps_2_slinear = gaps_2.interpolate(method="slinear")
gaps_2_slinear.name = "slinear"

fig_pandas, pandas_gaps_1_ax, pandas_gaps_2_ax = plot(original=original,
    gaps_1_method_1=gaps_1_spline,
    gaps_1_method_2=gaps_1_slinear,
    gaps_2_method_1=gaps_2_spline,
    gaps_2_method_2=gaps_2_slinear)
fig_pandas.savefig("pandas_interpolation.png")
```
